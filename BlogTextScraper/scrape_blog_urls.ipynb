```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import json
import os

base_url = "https://futurice.com/blog?page="
```

```python
#Scrapes all blog links form the "page_num":th page
def scrape_page(page_num):
    urls = []
    base_url = "https://futurice.com/blog?page="
    
    r = requests.get(base_url + str(page_num))

    if r.status_code//100 != 2:
        print("ERROR WHILE READING WEBPAGE")
        return List()
    
    soup = BeautifulSoup(r.text, 'html.parser')
    
    for item in soup.body.main.find_all("a"):
        if item.get("href"):
            urls.append(item.get("href"))
        
    return urls

# Scrapes all the pages
def scrape_blog_urls(verbose = False):
    scrape_more = True
    index = 1
    url_count= 0
    urls = []

    while scrape_more:

        urls.extend(scrape_page(index))
        if len(urls) == url_count:
            scrape_more = False

        if verbose: {print(str(index) + ": " + str(len(urls)))}

        index += 1
        url_count = len(urls)


    if verbose: {print("Finnished")}
    return urls


# Returns path to the upper directory
def upper_directory(file_name):
    return str(os.path.dirname(os.path.dirname(os.path.abspath(file_name))))

# Writes the urls to file
def write_to_file(urls):
    file_path = upper_directory("scrape_blog_urls") + "/data/blog_urls_all.csv"
    with open(file_path, "w", encoding='utf-8') as file:
        file.write("url\n")
        for item in urls:
            file.write(item + "\n")
```

```python
# Calling functions to get things scrapped
urls = scrape_blog_urls()
write_to_file(urls)
```
